```python
import numpy as np
import matplotlib.pyplot as plt
import math

def PlotFunction(begin_x, end_x, n_points, func):
    X = np.linspace(begin_x, end_x, n_points, endpoint=True)
    Y = map(lambda x:func(x), X)
    plt.plot(X, Y)
    plt.show()
    

PlotFunction(0.01, 30, 100, lambda x:3.0/(1+math.exp(x/3.0))+0.5)
```

```python
# log-sigmoid

import math

# sigmoid beyond 0.5
def logsigmoid(x):
    return -math.log(1/(1+math.exp(-x)))

# sigmoid beyond 0.5
def sigmoid(x):
    return 1/(1+math.exp(-x))

pydev.function_curve(-10, 10, 0.1, logsigmoid)

print logsigmoid(0)
print sigmoid(0)

print logsigmoid(10)
print sigmoid(10)
```
